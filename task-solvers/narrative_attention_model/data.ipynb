```python
import pandas as pd

def process_labels(df):
    """
    Process a DataFrame by:
    1. Exploding the 'labels' column based on semicolon (;) delimiter
    2. Splitting each label at colon (:) to create narrative hierarchy columns
    3. Replacing empty subnarrative and subsubnarrative values with "Other"
    
    Parameters:
    df (pandas.DataFrame): DataFrame containing a 'labels' column
    
    Returns:
    pandas.DataFrame: Processed DataFrame with new columns
    """
    # Create a copy to avoid modifying the original DataFrame
    result_df = df.copy()
    
    # Step 1: Explode the 'labels' column based on semicolon separator
    result_df = result_df.assign(labels=result_df['labels'].str.split(';')).explode('labels')
    
    # Step 2: Split each label into narrative, subnarrative, and subsubnarrative
    split_labels = result_df['labels'].str.split(':', expand=True)
    
    # Ensure we have 3 columns even if some rows don't have all levels
    if len(split_labels.columns) < 3:
        for i in range(len(split_labels.columns), 3):
            split_labels[i] = ""
    
    # Rename the columns
    split_labels.columns = ['narrative', 'subnarrative', 'subsubnarrative']
    
    # Add these columns to our result DataFrame
    result_df = pd.concat([result_df, split_labels], axis=1)
    
    # Step 3: Replace empty subnarrative and subsubnarrative with "Other"
    result_df['subnarrative'] = result_df['subnarrative'].replace('', 'Other')
    result_df['subsubnarrative'] = result_df['subsubnarrative'].replace('', 'Other')
    
    # Clean up any whitespace
    for col in ['narrative', 'subnarrative', 'subsubnarrative']:
        result_df[col] = result_df[col].str.strip()
    
    return result_df
```
