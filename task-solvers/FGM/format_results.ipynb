```python
# Assuming you already have the labels column as lists
def process_labels(labels_list):
    # If list is empty, return "Other"
    if not labels_list:
        return "Other"
    # If list has 2 or more elements, join with ";"
    elif len(labels_list) >= 2:
        return ";".join(str(label) for label in labels_list)
    # If list has 1 element, return it as string
    else:
        return str(labels_list[0])

# Apply to your DataFrame
result_df['labels'] = result_df['labels'].apply(process_labels)

# Example how it would work:
# Empty list -> "Other"
# ['label1'] -> "label1"
# ['label1', 'label2'] -> "label1;label2"
# ['label1', 'label2', 'label3'] -> "label1;label2;label3"
```

```python
def create_narrative(label_str):
    # If label is "Other", return as is
    if label_str == "Other":
        return "Other"
    
    # Split multiple labels if they exist
    labels = label_str.split(';')
    
    # Process each label to get only top_cat:mid_cat
    narratives = []
    for label in labels:
        # Split the hierarchical levels
        parts = label.split(':')
        if len(parts) >= 2:  # Ensure we have at least top and mid category
            narratives.append(f"{parts[0]}:{parts[1]}")
    
    # Join multiple narratives with semicolon
    return ';'.join(narratives) if narratives else "Other"

# Apply to DataFrame
result_df['narrative'] = result_df['labels'].apply(create_narrative)

# Example of how it works:
# Input label: "tech:software:backend" -> "tech:software"
# Input label: "tech:software:backend;health:medical:cardio" -> "tech:software;health:medical"
# Input label: "Other" -> "Other"
```

```python
import pandas as pd

def get_labels(df):
    """
    Convert one-hot encoded labels to a list format.
    Args:
        df (pd.DataFrame): DataFrame with 'text' and label columns
    Returns:
        pd.Series: Series containing lists of labels for each row
    """
    # Get all columns except 'text'
    label_cols = df.drop('text', axis=1)
    
    # Convert to list of labels where value is 1
    labels = label_cols.apply(lambda row: row[row == 1].index.tolist(), axis=1)
    return labels

def process_labels(labels_series):
    """
    Process label lists according to specific rules:
    - Empty lists become "Other"
    - Multiple labels are joined with ";"
    - All elements are converted to strings
    
    Args:
        labels_series (pd.Series): Series containing lists of labels
    Returns:
        pd.Series: Processed labels as strings
    """
    def process_single_label(labels_list):
        if not labels_list:
            return "Other"
        elif len(labels_list) >= 2:
            return ";".join(str(label) for label in labels_list)
        else:
            return str(labels_list[0])
    
    return labels_series.apply(process_single_label)

def create_narrative(labels_series):
    """
    Create narratives from hierarchical labels.
    Format: top_cat:mid_cat from top_cat:mid_cat:low_cat
    
    Args:
        labels_series (pd.Series): Series containing label strings
    Returns:
        pd.Series: Narratives derived from labels
    """
    def process_single_narrative(label_str):
        # Handle "Other" case
        if label_str == "Other":
            return "Other"
        
        # Split multiple labels if they exist
        labels = label_str.split(';')
        
        # Process each label to get only top_cat:mid_cat
        narratives = []
        for label in labels:
            parts = label.split(':')
            if len(parts) >= 2:  # Ensure we have at least top and mid category
                narratives.append(f"{parts[0]}:{parts[1]}")
        
        return ';'.join(narratives) if narratives else "Other"
    
    return labels_series.apply(process_single_narrative)

def process_dataframe(df):
    """
    Apply all three transformations to the DataFrame in sequence.
    
    Args:
        df (pd.DataFrame): Input DataFrame with 'text' and one-hot encoded label columns
    Returns:
        pd.DataFrame: Processed DataFrame with new 'labels' and 'narrative' columns
    """
    # Make a copy to avoid modifying the original
    result_df = df.copy()
    
    # Apply transformations in sequence
    result_df['labels'] = get_labels(df)
    result_df['subnarrative'] = process_labels(result_df['labels'])
    result_df['narrative'] = create_narrative(result_df['subnarrative'])
    
    # Keep only necessary columns
    result_df = result_df[['file','narrative', 'subnarrative']]
    
    return result_df

# Example usage:
    # Example DataFrame
df =  pd.read_csv("EN_results_onehot.csv")
    # Process the DataFrame
result = process_dataframe(df)
print("\nProcessed DataFrame:")
result
```

```python
langs = ["EN", "HI", "BG", "RU", "PT"]
for lang in langs:
    df = pd.read_csv(f"{lang}_results_onehot.csv")
    result = process_dataframe(df)
    result.to_csv(f"{lang}_results_formatted.txt", index=False, sep="\t", header=False)
```
